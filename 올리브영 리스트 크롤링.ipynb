```python
import requests
from selenium import webdriver
import time
```

```python
from bs4 import BeautifulSoup

#페이지 자르기(pageIdx 기준)
url1 = "https://www.oliveyoung.co.kr/store/display/getMCategoryList.do?dispCatNo=100000100020006&fltDispCatNo=&prdSort=01&pageIdx="
url2 ="&rowsPerPage=24&searchTypeSort=btn_thumb&plusButtonFlag=N&isLoginCnt=0&aShowCnt=0&bShowCnt=0&cShowCnt=0&trackingCd=Cat100000100020006_MID"

code_list = []

for a in range(1,11):
    url = url1 + str(a) + url2
    res = requests.get(url)
    html=res.text
    soup=BeautifulSoup(html, 'html.parser')

    for i in range(8, 14):
        #8-13까지 변경하면서
        for j in range(1, 5):
            if j == 1:
                codes=soup.select("#Contents ul:nth-child(%s) li.flag" %i)
            else:
                codes=soup.select("#Contents ul:nth-child(%s) li:nth-child(%s)" %(i,j))

            import re 
            for code in codes:
                num = codes[0].select_one('a[href]')
                lists = re.findall(r'\d{12}', num['href'])
            code_list.append(lists[0])
            print(a,"페이지", i, "번줄", j, "번째", lists[0])
code_list
```
