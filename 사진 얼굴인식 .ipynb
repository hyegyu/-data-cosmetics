```python
from mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt

image_path = 'sum.jpg'
img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
detector = MTCNN()
detections = detector.detect_faces(img)
print(detections)

img_with_dets = img.copy()
min_conf = 0.9
for det in detections:
    if det['confidence'] >= min_conf:
        x, y, width, height = det['box']
        keypoints = det['keypoints']
        cv2.rectangle(img_with_dets, (x,y), (x+width,y+height), (0,155,255),20)
        cv2.circle(img_with_dets, (keypoints['left_eye']), 2, (0,155,255), 20)
        cv2.circle(img_with_dets, (keypoints['right_eye']), 2, (0,155,255), 20)
        cv2.circle(img_with_dets, (keypoints['nose']), 2, (0,155,255), 20)
        cv2.circle(img_with_dets, (keypoints['mouth_left']), 2, (0,155,255), 20)
        cv2.circle(img_with_dets, (keypoints['mouth_right']), 2, (0,155,255), 20)
        
plt.figure(figsize = (10,10))
plt.imshow(img_with_dets)
plt.axis('off')
```
